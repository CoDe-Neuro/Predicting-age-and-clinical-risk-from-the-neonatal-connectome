```python
import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import numpy as np
import os
import scipy.io as scp
import copy
import random
from sklearn.model_selection import StratifiedKFold
import scipy
import math
import pickle
import pandas as pd

from sklearn.linear_model import LinearRegression
```

```python
def normalize_vector(vector, min_, max_):
    return (vector-min_)/(max_- min_)

def Vectorize_Connectivity_Matrix(Original_Matrix):
    N_Subjects = Original_Matrix.shape[2]
    N_Brain_Regions = 90
    #Lower Triangle n_elements --> (90*90/2) - 90/2 = 4005
    N_Elements_In_Vector = 4005 

    Vectorized_Matrix = np.zeros((N_Subjects, N_Elements_In_Vector)) 
  
    for n in range(N_Subjects):
        Counter = 0
        for i in range(N_Brain_Regions):
            for j in range(N_Brain_Regions):
                if i>j:
                    Vectorized_Matrix[n, Counter] = Original_Matrix[i,j,n]
                    Counter +=1
                  
    return Vectorized_Matrix

def flatten_list(list_):
    new_list = []
    for i in list_:
        for j in i:
            new_list.append(j)
    return new_list

def Get_Mean_Absolute_Error(ytest, ypred):
    MAE_ = np.zeros(len(ytest), )

    for i in range(len(ytest)):
        MAE_[i] = np.abs(ytest[i]-ypred[i])
    return np.mean(MAE_)
    
def reconstruct_Connectome_From_Vector(Connectivity_Vector, nx = 90, ny = 90, return_indexes = False):
    reconstructed_matrix = np.zeros((90,90))
    connection_indexes = []
    k=0
    for i in range(nx):
        for j in range(ny):
            if i>j:
                reconstructed_matrix[i,j] = Connectivity_Vector[k]
                reconstructed_matrix[j,i] = Connectivity_Vector[k]
                connection_indexes.append([i,j])
                k = k+1
                
    if return_indexes == True:       
        return reconstructed_matrix, connection_indexes
    else:
        return reconstructed_matrix
    
def get_N_Most_Important_Connections(feature_vector, connection_indexes, connection_names, N):
    
    sorted_feature_importance = np.sort(feature_vector)[::-1]
    sorted_feature_importance_indexes = np.argsort(feature_vector)[::-1]
    most_important_features = []
    for i in range(N):
        ij = connection_indexes[sorted_feature_importance_indexes[i]]
        importance = sorted_feature_importance[i]
        connection_name = [connection_names[ij[0]], connection_names[ij[1]]]
        most_important_features.append([ij, connection_name, importance])
        
    return most_important_features

def get_N_Most_Important_Regions(alterations_connectome, connection_names, N_regions):
    
    Region_Weight = np.zeros((90,))
    
    for i in range(alterations_connectome.shape[0]):
        for j in range(alterations_connectome.shape[1]):
            Region_Weight[i] += alterations_connectome[i,j]
            
    Sorted_Region_Weights = np.sort(Region_Weight)[::-1]
    Sorted_Region_Weights_Indexes = np.argsort(Region_Weight)[::-1]
        
    Outputed_Regions = []
    
    for n in range(N_regions):
        Outputed_Regions.append(connection_names[Sorted_Region_Weights_Indexes[n]])

    return Outputed_Regions

def print_Connections_And_Regions(connection_list, regions_list, model):
    print('\n', 'model: ', model)
    print('\tConnections:')
    for i in range(len(connection_list)):
        print('\t\t', connection_list[i][1][0], '||||', connection_list[i][1][1], ':' , connection_list[i][2])
    print('\tRegions')
    for i in range(len(regions_list)):
        print('\t\t', regions_list[i])
        
def Write_CSV(connections_ranking, regions_ranking, path):
    connection_1 = []
    connection_2 = []
    
    for i in range(len(connections_ranking)):
        connection_1.append(connections_ranking[i][1][0])
        connection_2.append(connections_ranking[i][1][1])
        
    df = pd.DataFrame({'Connection 1': connection_2,
                   'Connection 2': connection_1,
                   'Region': regions_ranking})
    df.to_csv(path, index=False)
    
def open_file(path_to_file):
    if path_to_file.endswith('pkl'): #This is a variable 
        with open(path_to_file, 'rb') as f:
            return pickle.load(f)
    elif path_to_file.endswith('npz'): #This is also a variable
        return np.load(path_to_file, allow_pickle=True)['arr_0']
    
def transform_vector(vector, edge_to_alter, transformation):
    if transformation == 'Zero':
        transformed_vector = copy.deepcopy(vector)
        transformed_vector[edge_to_alter] = 0
    return transformed_vector

def load_RF_Model(path):
    with open(path, 'rb') as f:
        model = pickle.load(f)
    return model
```

```python
#Load data from matlab struct - turn into np array, and verify shape is correct

Path_to_Connectivity = os.path.abspath('Data/NewStructuralConnectivity.mat')
Data = scp.loadmat(Path_to_Connectivity)

PMA = np.array(Data.get('pma'))
GA = np.array(Data.get('ga'))
Sex = np.transpose(np.array(Data.get('sex')))
Structural = np.array(Data.get('SCmu'))
N = Structural.shape[2]
Names = Data.get('sub')
Subject_ID = [x.tolist()[0][4:] for x in Names[0]]

print('Shape of sex : ' , Sex.shape)
print('Shape of GA : ' , GA.shape)
print('Shape of PMA : ', PMA.shape)
print('Shape of Structural Matrix : ', Structural.shape)

X = Vectorize_Connectivity_Matrix(Structural)

Subject_Dictionary = {}
for i in range(N):
    Subject_Dictionary[Subject_ID[i]] = {}
    Subject_Dictionary[Subject_ID[i]]['GA']= float(GA[i][0])
    Subject_Dictionary[Subject_ID[i]]['PMA'] = float(PMA[i][0])
    Subject_Dictionary[Subject_ID[i]]['Sex'] = Sex[0][i]
    Subject_Dictionary[Subject_ID[i]]['Connectome'] = X[i, :]
    
Connection_Names = list(pd.read_csv('Data/Connection Names.csv')['Name'])
```

# Starting PMA sensitivity analysis...

```python
N_Folds = 5
N_Edges = 4005

test_folded_subjects = open_file('PMA_Files/Test_Fold_Subjects_PMA.pkl')
normalization_parameters = open_file('PMA_Files/RF/Normalization_Parameters_PMA_RF.pkl')
    
alteration_deltas = []
    
for fold_n in range(N_Folds):
    
    fold_test_subjects = test_folded_subjects[fold_n]
    fold_model = load_RF_Model(str('PMA_Files/RF/PMA_RF_Model') + str(fold_n) +  '.pkl')
    
    fold_alteration_deltas = []
    
    print('Working on fold:', fold_n, '...')
    for subject in fold_test_subjects:
        subject_x = Subject_Dictionary[subject]['Connectome']
        normalized_subject_x = normalize_vector(np.array(subject_x), \
                                normalization_parameters[fold_n][0], normalization_parameters[fold_n][1])
        
        altered_vectors = np.zeros((N_Edges, N_Edges))
        
        for edge in range(N_Edges):
            altered_vectors[edge, :] = transform_vector(vector = normalized_subject_x, \
                                        edge_to_alter = edge, transformation = 'Zero')
                
        altered_prediction = fold_model.predict(altered_vectors)
        actual_prediction = fold_model.predict(np.array(normalized_subject_x).reshape(1, 4005))
        fold_alteration_deltas.append(np.abs(altered_prediction - actual_prediction))
            
    alteration_deltas.append(fold_alteration_deltas)
        
```

# Obtain most important features

```python
plt.figure(figsize = (18,12))

sum_of_alterations = [0]*4005

for fold in range(5):
    fold_alteration_mean = np.zeros((4005,))
    
    fold_alterations_array = np.array(alteration_deltas[fold])
    for i in range(4005):
        fold_alteration_mean[i] = np.mean(fold_alterations_array[:,i])
        
    sum_of_alterations += fold_alteration_mean    
    
    fold_means_connectome, fold_connection_indexes = \
            reconstruct_Connectome_From_Vector(fold_alteration_mean, return_indexes = True)

    N_most_important_connections_fold = get_N_Most_Important_Connections(fold_alteration_mean, \
                                                fold_connection_indexes, Connection_Names, N = 10)
    
    N_most_important_regions_fold = get_N_Most_Important_Regions(fold_means_connectome, Connection_Names, N_regions = 10)
    
    
    print_Connections_And_Regions(N_most_important_connections_fold, N_most_important_regions_fold, model = fold)
    plt.subplot(2,3,(fold+1))
    plt.imshow(fold_means_connectome)
    plt.title(str('PMA RF model N' + str(fold+1)))
    
feature_matrix_sum, connection_indexes_sum = reconstruct_Connectome_From_Vector(sum_of_alterations, return_indexes = True)

N_most_important_connections_sum = get_N_Most_Important_Connections(sum_of_alterations, connection_indexes_sum, Connection_Names, N = 10)
N_most_important_regions_sum = get_N_Most_Important_Regions(feature_matrix_sum, Connection_Names, N_regions = 10)
print_Connections_And_Regions(N_most_important_connections_sum, N_most_important_regions_sum, model = 'Combination')
# plt.subplot(2,3,6)
# plt.imshow(feature_matrix_sum)
# plt.title(str('PMA RF model Combination'))
# Write_CSV(connections_ranking =N_most_important_connections_sum , regions_ranking = N_most_important_regions_sum, \
#                                                           path = 'PMA_RF.csv')

```

## Show main relevance map

```python
plt.figure(figsize = (10,10))
plt.imshow(feature_matrix_sum)
plt.title(str('Connection Relevance Map: \nPMA prediction with RF'), fontsize = 28)
plt.xticks(fontsize = 22)
plt.yticks(fontsize = 22)
plt.savefig('Map_RF_PMA.tiff', dpi = 350)
```
